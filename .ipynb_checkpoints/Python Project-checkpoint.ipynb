### General Descriptive Questions
1. Distribution of Claims: What percentage of travel insurance policies result in claims?
2. Popular Destinations: What are the most common travel destinations for insured individuals?
3. Age Distribution: What is the age distribution of insured individuals?

```python
# 1.Distribution of Claims: What percentage of travel insurance policies result in claims?

mask_claim_yes = df['Claim'] == 'Yes'
travel_claimed = df.loc[mask_claim_yes,'Claim'].count()
total_travel = len(df['Claim'])

pct_claimed = round(travel_claimed/total_travel * 100, 2)
print('Distribution of Claims:',pct_claimed,'%')
```

```python
# Popular Destinations for women:

mask_f = df['Gender'] == 'F'

df.loc[mask_f,'Destination'].value_counts().head()
```

### Product Analysis
1. Product Popularity: Which travel insurance products are the most popular?
2. Claims by Product: Which travel insurance products have the highest and lowest claim rates?
3. Duration of Travel: How does the duration of travel vary across different insurance products?

```python
# 1. Product Popularity: Which travel insurance products are the most popular?
df['Product Name'].value_counts().head(5)
```

```python
#2. Claims by Product: Which travel insurance products have the highest and lowest claim rates?
temp_df = df.pivot_table(index='Product Name', 
                         columns='Claim',
                         values='Agency', 
                         aggfunc='count').fillna(0)
print(temp_df)
print('\n-----------------------------------------------------')

product_name_max = temp_df.loc[:,'Yes'].idxmax()
product_value_max = temp_df.loc[:,'Yes'].max()
print('\nMost insurance product claimed:',product_name_max,'\nwith the amount of:',product_value_max)

product_value_min = temp_df.loc[:,'Yes'].min()
print('\nThe least amount of insurance product claimed is:',product_value_min)
mask = temp_df['Yes'] == product_value_min
print('The list of insurance products with that amount are:')
for i in temp_df[mask].index:
    print(i)
```

```python
#3. Duration of Travel: How does the duration of travel vary across different insurance products?
g_Product_Name = df.groupby('Product Name')

g_Product_Name['Duration'].mean().sort_values(ascending=False)
```

### Demographic Analysis
1. Gender Analysis: Are there differences in claim rates between male and female insured individuals?
2. Age and Claims: Is there a correlation between the age of the insured individuals and the likelihood of making a claim?
3. Age and Travel Destination: Do travel destinations vary significantly by age group?

```python
#1. Gender Analysis: Are there differences in claim rates between male and female insured individuals?
df.pivot_table(index='Gender', 
               columns='Claim',
               values='Agency', 
               aggfunc='count')
```

```python
#2. Age and Claims: Is there a correlation between the age of the insured individuals and the likelihood of making a claim?
import scipy.stats as stats

mask_claim_yes = df['Claim'] == 'Yes'
mask_claim_no = df['Claim'] == 'No'

# Creating a copy of the column 'Claim' and turn it to a column with binary values 
df['claim_code'] = df['Claim']
df.loc[mask_claim_yes,'claim_code'] = 1
df.loc[mask_claim_no,'claim_code'] = 0

# Calculate point biserial correlation
correlation, p_value = stats.pointbiserialr(df['Age'], df['claim_code'])

# Display the result
print("Point Biserial Correlation:", correlation)
print("P-value:", p_value)
print('\nFrom that we can understand that there is almost no correlation')
```

# Graphs

```python
#General Descriptive Questions - Q1

# Calculate the total number of policies and the number of claimed policies
total_policies = df.shape[0]
claimed_policies = df[df['Claim'] == 'Yes'].shape[0]

# Calculate the percentage of claimed and unclaimed policies
claimed_percentage = (claimed_policies / total_policies) * 100
unclaimed_percentage = 100 - claimed_percentage

# Prepare the data for the pie chart
labels = ['Claimed', 'Unclaimed']
sizes = [claimed_percentage, unclaimed_percentage]
colors = ['#ff9999','#66b3ff']
explode = (0.1, 0)  # explode the 'Claimed' slice

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', shadow=True, startangle=140)
plt.title('Percentage of Claimed vs Unclaimed Policies')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#General Descriptive Questions - Q2

# Calculate the top 5 destinations
top_destinations = df['Destination'].value_counts().head()

# Plot the bar chart
plt.figure(figsize=(10, 6))
bars = top_destinations.plot(kind='bar', color='skyblue')
plt.title('Top 5 Travel Insurance Destinations')
plt.xlabel('Destination')
plt.ylabel('Number of Policies')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Add value labels on top of the bars
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 0.05,
             round(bar.get_height(), 2),
             ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
#General Descriptive Questions - Q3

# Calculate the value counts for each age group
age_group_counts = df['Age_group'].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(10, 6))
bars = age_group_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Policies')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Add value labels on top of the bars
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 0.05,
             int(bar.get_height()),
             ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
# Agency and Distribution Channel Analysis - Q1

# Calculation of the number of policies by agency and sorting in descending order
agency_counts = df['Agency'].value_counts().sort_values(ascending=False).head()

# Create a bar graph with narrower columns
plt.figure(figsize=(12, 8))
ax = sns.barplot(x=agency_counts.index, y=agency_counts.values, palette='viridis', width=0.5)

# Adding numeric values above the columns
for i, (agency, count) in enumerate(zip(agency_counts.index, agency_counts.values)):
    ax.text(i, count + 0.2, f'{count}', ha='center', va='bottom', color='black', fontsize=12, weight='bold')

# Adding titles and axes
plt.title('Top 5 Agencies by Number of Policies Sold')
plt.xlabel('Agency')
plt.ylabel('Number of Policies')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Product Analysis - Q1

# Introducing the five most popular products
top_products = df['Product Name'].value_counts().head(5)

# Creating a bar graph
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=top_products.index, y=top_products.values, palette='viridis', width=0.5)

# Adding numeric values above the columns
for i, (product, count) in enumerate(zip(top_products.index, top_products.values)):
    ax.text(i, count + 0.5, f'{count}', ha='center', va='bottom', color='black', fontsize=12, weight='bold')

# Adding titles and axes
plt.title('Top 5 Product Types')
plt.xlabel('Product Name')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Product Analysis - Q3

# Calculation of the average travel time according to the type of insurance and sorted in descending order
mean_duration = df.groupby('Product Name')['Duration'].mean().sort_values(ascending=False)

# Choosing the top five insurance products
top5_mean_duration = mean_duration.head(5)

# Creating a bar graph
plt.figure(figsize=(12, 8))
ax = sns.barplot(x=top5_mean_duration.index, y=top5_mean_duration.values, palette='viridis', width=0.5)

# Adding numeric values above the columns
for i, (product, duration) in enumerate(zip(top5_mean_duration.index, top5_mean_duration.values)):
    ax.text(i, duration + 0.2, f'{duration:.2f}', ha='center', va='bottom', color='black', fontsize=12, weight='bold')

# Adding titles and axes
plt.title('Average Duration of Travel by Top 5 Insurance Products')
plt.xlabel('Product Name')
plt.ylabel('Average Duration (Days)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Demographic Analysis - Q1

# Create the pivot table
pivot_df = df.pivot_table(index='Gender', 
                          columns='Claim', 
                          values='Agency', 
                          aggfunc='count')

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Get the data from the pivot table
x = np.arange(len(pivot_df.index))
width = 0.35
multiplier = 0

# Plot each 'Claim' category as a group of bars
for column in pivot_df.columns:
    offset = width * multiplier
    rects = ax.bar(x + offset, pivot_df[column], width, label=column)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Customize the plot
ax.set_ylabel('Count')
ax.set_title('Agency Count by Gender and Claim')
ax.set_xticks(x + width, pivot_df.index)
ax.legend(loc='best', title='Claim')
ax.set_ylim(0, pivot_df.values.max() * 1.1)  # Set y-axis limit with some padding

# Add some spacing between subplot elements
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Demographic Analysis - Q2

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['claim_code'], alpha=0.5)
plt.title('Scatter plot of Age vs Claim Code')
plt.xlabel('Age')
plt.ylabel('Claim Code')
plt.grid(True)
plt.show()
```

```python
# Demographic Analysis - Q3

# Create the pivot table
pivot_table = df.pivot_table(index='Destination', 
                             columns='Age_group',
                             values='Agency', 
                             aggfunc='count').sort_values(['0-20', '21-40', '41-60', '61-80', '81-100', '101-120'], 
                                                          ascending=False).head(10)

# Plot the pivot table
pivot_table.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='viridis')

# Set the plot title and labels
plt.title('Top 10 Destinations by Age Group and Agency Count')
plt.xlabel('Destination')
plt.ylabel('Agency Count')
plt.legend(title='Age Group')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
